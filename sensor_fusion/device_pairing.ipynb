#  time lagged cross correlation 
### TLCC (time lagged cross correlation) is measured by incrementally shifting one time series and repeatedly calculating the correlation between two signals. The offset is the lag with the highest correleration. The smaller the peak offset, the more synchronized the two time series. If the peak offset is too large, the two series are said to be not synchronized, thus from a malicious attacker device. 

```python
def crosscorr(datax, datay, lag=0, wrap=False):
    """ Lag-N cross correlation. 
    Shifted data filled with NaNs 
    
    Parameters
    ----------
    lag : int, default 0
    datax, datay : pandas.Series objects of equal length
    Returns
    ----------
    crosscorr : float
    """
    if wrap:
        shiftedy = datay.shift(lag)
        shiftedy.iloc[:lag] = datay.iloc[-lag:].values
        return datax.corr(shiftedy)
    return datax.corr(datay.shift(lag))



```

```python
def plot_offset_graph(d1, d2):
    max_lag = 150
    rs = [crosscorr(d1,d2, lag) for lag in range(-int(max_lag),int(max_lag+1))]
    offset = np.ceil(len(rs)/2)-np.argmax(rs)
    f,ax=plt.subplots(figsize=(14,3))
    ax.plot(rs)
    ax.axvline(np.ceil(len(rs)/2),color='k',linestyle='--',label='Center')
    ax.axvline(np.argmax(rs),color='r',linestyle='--',label='Peak synchrony')
    ax.set(title=f'Offset = {offset} frames\nS1 leads <> S2 leads',ylim=[0,1],xlim=[0,301], xlabel='Offset',ylabel='Pearson r')
    ax.set_xticks([0, 50, 100, 151, 201, 251, 301])
    ax.set_xticklabels([-150, -100, -50, 0, 50, 100, 150]);
    plt.legend()
    plt.show()
    return offset, max(rs) #return lag with highest correlation
```

## Prediction 
### if the lag is greater than 40% of the maximum lag, we label it as negative (from a malicious attacker device) 

```python
def plot_offset_graph(d1, d2):
    max_lag = 150
    rs = [crosscorr(d1,d2, lag) for lag in range(-int(max_lag),int(max_lag+1))]
    offset = np.ceil(len(rs)/2)-np.argmax(rs)
    f,ax=plt.subplots(figsize=(14,3))
    ax.plot(rs)
    ax.axvline(np.ceil(len(rs)/2),color='k',linestyle='--',label='Center')
    ax.axvline(np.argmax(rs),color='r',linestyle='--',label='Peak synchrony')
    ax.set(title=f'Offset = {offset} frames\nS1 leads <> S2 leads',ylim=[0,1],xlim=[0,301], xlabel='Offset',ylabel='Pearson r')
    ax.set_xticks([0, 50, 100, 151, 201, 251, 301])
    ax.set_xticklabels([-150, -100, -50, 0, 50, 100, 150]);
    plt.legend()
    plt.show()
    return offset, max(rs),rs #return lag with highest correlation
```

```python
def pred(d1, d2):
    
    max_lag = 150
    rs = [crosscorr(d1,d2, lag) for lag in range(-int(max_lag),int(max_lag+1))]
    offset = np.ceil(len(rs)/2)-np.argmax(rs)
    prediction = 0
    if (offset > 0.4*max_lag or offset < -0.4*max_lag):
        prediction = 1
    return prediction
```

## Simulate negative testing data 

```python
def simulate_temperature(size = 1000, num_samples = 1000):
    temperatures = []
    for i in range(num_samples):
        simulated_temp = []
        for i in range(size):
            temp = int((truncnorm((65 - 75) / 2, (80 - 75) / 2, loc=75, scale=2).rvs()))
            simulated_temp.append(temp)
        temperatures.append(simulated_temp)
    return temperatures
```

```python
def test(simulated_temperatures):
    t = 0
    for i in range(len(simulated_temperatures)):  
        if (pred( pd.Series(simulated_temperatures[i]), df['temperature_sensorA'][100:1100].astype(float).reset_index(drop = True) ) == 1):
            t = t + 1
    return t /len(simulated_temperatures)
```
